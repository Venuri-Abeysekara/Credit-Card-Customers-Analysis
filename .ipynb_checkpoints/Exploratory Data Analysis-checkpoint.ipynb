```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, confusion_matrix, 
                             roc_auc_score, roc_curve, accuracy_score)
from sklearn.cluster import KMeans
import warnings

from IPython.display import Markdown

warnings.filterwarnings('ignore')

#visualization style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

display(Markdown("**VARIABLE IMPORTANCE ANALYSIS**"))
display(Markdown(
    "Identifying which variables best predict customer churn and "
    "build models to identify at risk customers before they leave"
))
```

```python
df = pd.read_csv('BankChurners.csv')

print(f" Dataset loaded successfully!")
print(f"   Total customers: {len(df):,}")
print(f"   Total features: {len(df.columns)}")

# Create binary churn variable
df['Churn'] = (df['Attrition_Flag'] == 'Attrited Customer').astype(int)

churn_count = df['Churn'].sum()
churn_rate = (churn_count / len(df)) * 100

print(f"\n Churn Overview:")
print(f"   Churned customers: {churn_count:,} ({churn_rate:.2f}%)")
print(f"   Retained customers: {len(df) - churn_count:,} ({100-churn_rate:.2f}%)")
```

```python
# Analyzing correlation of all variables with churn
display(Markdown("**VARIABLE CORRELATION ANALYSIS**"))

# All numerical variables
numerical_vars = ['Customer_Age', 'Dependent_count', 'Months_on_book',
                  'Total_Relationship_Count', 'Months_Inactive_12_mon',
                  'Contacts_Count_12_mon', 'Credit_Limit', 
                  'Total_Revolving_Bal', 'Avg_Open_To_Buy',
                  'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt', 
                  'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1',
                  'Avg_Utilization_Ratio', 'Churn']

# Correlations
correlation = df[numerical_vars].corr()['Churn'].sort_values(ascending=False)

display(Markdown(
    "Correlation with Churn (sorted by strength):"
))

print(f"{'Variable':<30} {'Correlation':<15} {'Strength'}")
print("-" * 80)

for var, corr in correlation.items():
    if var != 'Churn':
        # Determining strength
        abs_corr = abs(corr)
        if abs_corr > 0.3:
            strength = "⭐ STRONG"
        elif abs_corr > 0.1:
            strength = "✅ MODERATE"
        else:
            strength = "⚠️  WEAK"
        
        print(f"{var:<30} {corr:>+.4f}          {strength}")

print("\n")

# Identifying top predictors
top_predictors = correlation.drop('Churn').abs().nlargest(5)
display(Markdown("\n TOP 5 CHURN PREDICTORS:"))
for i, (var, corr) in enumerate(top_predictors.items(), 1):
    print(f"   {i}. {var}: {correlation[var]:+.4f}")
```

```python
display(Markdown("\n **GENERATING CORRELATION HEATMAP**"))

plt.figure(figsize=(16, 16))

# Full correlation matrix
corr_matrix = df[numerical_vars].corr()

# Heatmap
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8},
            center=0, vmin=-1, vmax=1)

plt.title('Correlation Matrix - All Variables vs Churn', 
          fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

#Reading guid
display(Markdown("\n **Reading Guid:**"))
display(Markdown("   Red = Strong positive correlation (as one ↑, other ↑)"))
display(Markdown("   Blue = Strong negative correlation (as one ↑, other ↓)"))
display(Markdown("   White = No correlation"))
```

```python
# Top predictors
display(Markdown("\n **TOP CHURN PREDICTORS** \n"))

display(Markdown("\n The box plots show the distribution of each variable for churned vs existing customers"))

print("\n")

top_5_vars = correlation.drop('Churn').abs().nlargest(5).index.tolist()

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.flatten()

for idx, var in enumerate(top_5_vars):
    # Box plot
    df.boxplot(column=var, by='Attrition_Flag', ax=axes[idx])
    axes[idx].set_title(f'{var}\n(Correlation: {correlation[var]:+.3f})', 
                        fontweight='bold')
    axes[idx].set_xlabel('Customer Status')
    axes[idx].set_ylabel(var)
    axes[idx].get_figure().suptitle('')
    
    # Calculate means
    churned_mean = df[df['Churn']==1][var].mean()
    existing_mean = df[df['Churn']==0][var].mean()
    
    # Add statistics
    stats_text = f"Churned: {churned_mean:.2f}\nExisting: {existing_mean:.2f}"
    axes[idx].text(0.02, 0.98, stats_text, transform=axes[idx].transAxes,
                   verticalalignment='top', bbox=dict(boxstyle='round', 
                   facecolor='wheat', alpha=0.5), fontsize=9)

# Remove extra subplot
axes[-1].axis('off')

plt.tight_layout()
plt.show()
```

```python
display(Markdown("\n **DETAILED DEMOGRAPHIC ANALYSIS**"))
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

print("\n")

# 1. Age Distribution by Churn Status
axes[0, 0].hist([df[df['Attrition_Flag']=='Existing Customer']['Customer_Age'],
                 df[df['Attrition_Flag']=='Attrited Customer']['Customer_Age']],
                label=['Retained', 'Churned'], bins=20, color=['green', 'red'], alpha=0.7)
axes[0, 0].set_xlabel('Age', fontweight='bold')
axes[0, 0].set_ylabel('Frequency', fontweight='bold')
axes[0, 0].set_title('Age Distribution by Churn Status', fontweight='bold', fontsize=13)
axes[0, 0].legend()
axes[0, 0].grid(alpha=0.3)

# 2. Gender Distribution
gender_churn = pd.crosstab(df['Gender'], df['Attrition_Flag'])
gender_churn.plot(kind='bar', ax=axes[0, 1], color=['green', 'red'], edgecolor='black')
axes[0, 1].set_xlabel('Gender', fontweight='bold')
axes[0, 1].set_ylabel('Count', fontweight='bold')
axes[0, 1].set_title('Churn by Gender', fontweight='bold', fontsize=13)
axes[0, 1].set_xticklabels(axes[0, 1].get_xticklabels(), rotation=0)
axes[0, 1].legend(['Retained', 'Churned'])
axes[0, 1].grid(alpha=0.3, axis='y')

# 3. Income Category
income_churn = pd.crosstab(df['Income_Category'], df['Attrition_Flag'], normalize='index') * 100
income_churn.plot(kind='bar', stacked=True, ax=axes[0, 2], color=['green', 'red'], edgecolor='black')
axes[0, 2].set_xlabel('Income Category', fontweight='bold')
axes[0, 2].set_ylabel('Percentage (%)', fontweight='bold')
axes[0, 2].set_title('Churn Rate by Income Category', fontweight='bold', fontsize=13)
axes[0, 2].set_xticklabels(axes[0, 2].get_xticklabels(), rotation=45, ha='right')
axes[0, 2].legend(['Retained', 'Churned'])
axes[0, 2].grid(alpha=0.3, axis='y')

# 4. Education Level
edu_churn = pd.crosstab(df['Education_Level'], df['Attrition_Flag'], normalize='index') * 100
edu_churn.plot(kind='bar', stacked=True, ax=axes[1, 0], color=['green', 'red'], edgecolor='black')
axes[1, 0].set_xlabel('Education Level', fontweight='bold')
axes[1, 0].set_ylabel('Percentage (%)', fontweight='bold')
axes[1, 0].set_title('Churn Rate by Education Level', fontweight='bold', fontsize=13)
axes[1, 0].set_xticklabels(axes[1, 0].get_xticklabels(), rotation=45, ha='right')
axes[1, 0].legend(['Retained', 'Churned'])
axes[1, 0].grid(alpha=0.3, axis='y')

# 5. Marital Status
marital_churn = pd.crosstab(df['Marital_Status'], df['Attrition_Flag'], normalize='index') * 100
marital_churn.plot(kind='bar', stacked=True, ax=axes[1, 1], color=['green', 'red'], edgecolor='black')
axes[1, 1].set_xlabel('Marital Status', fontweight='bold')
axes[1, 1].set_ylabel('Percentage (%)', fontweight='bold')
axes[1, 1].set_title('Churn Rate by Marital Status', fontweight='bold', fontsize=13)
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=0)
axes[1, 1].legend(['Retained', 'Churned'])
axes[1, 1].grid(alpha=0.3, axis='y')

# 6. Card Category
card_churn = pd.crosstab(df['Card_Category'], df['Attrition_Flag'], normalize='index') * 100
card_churn.plot(kind='bar', stacked=True, ax=axes[1, 2], color=['green', 'red'], edgecolor='black')
axes[1, 2].set_xlabel('Card Category', fontweight='bold')
axes[1, 2].set_ylabel('Percentage (%)', fontweight='bold')
axes[1, 2].set_title('Churn Rate by Card Category', fontweight='bold', fontsize=13)
axes[1, 2].set_xticklabels(axes[1, 2].get_xticklabels(), rotation=45, ha='right')
axes[1, 2].legend(['Retained', 'Churned'])
axes[1, 2].grid(alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print("\nKEY DEMOGRAPHIC INSIGHTS:\n")
print(f"   • Age: Churn is relatively consistent across age groups")
print(f"   • Gender: Minimal difference in churn rates between genders")
print(f"   • Income: Higher income categories show slightly lower churn")
print(f"   • Education: Graduate level shows moderate churn patterns")
print(f"   • Marital Status: Single customers show slightly higher churn")
print(f"   • Card Type: Blue card holders represent majority, consistent churn rate")
```

```python
display(Markdown("\n **TRANSACTION PATTERNS ANALYSIS**"))
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Transaction Count Distribution
axes[0, 0].hist([df[df['Attrition_Flag']=='Existing Customer']['Total_Trans_Ct'],
                 df[df['Attrition_Flag']=='Attrited Customer']['Total_Trans_Ct']],
                label=['Retained', 'Churned'], bins=30, color=['green', 'red'], alpha=0.7)
axes[0, 0].axvline(x=50, color='blue', linestyle='--', linewidth=2, label='Critical Threshold (50)')
axes[0, 0].set_xlabel('Total Transactions', fontweight='bold')
axes[0, 0].set_ylabel('Frequency', fontweight='bold')
axes[0, 0].set_title('Transaction Count Distribution', fontweight='bold', fontsize=13)
axes[0, 0].legend()
axes[0, 0].grid(alpha=0.3)

# 2. Transaction Amount Distribution
axes[0, 1].hist([df[df['Attrition_Flag']=='Existing Customer']['Total_Trans_Amt'],
                 df[df['Attrition_Flag']=='Attrited Customer']['Total_Trans_Amt']],
                label=['Retained', 'Churned'], bins=30, color=['green', 'red'], alpha=0.7)
axes[0, 1].set_xlabel('Total Transaction Amount ($)', fontweight='bold')
axes[0, 1].set_ylabel('Frequency', fontweight='bold')
axes[0, 1].set_title('Transaction Amount Distribution', fontweight='bold', fontsize=13)
axes[0, 1].legend()
axes[0, 1].grid(alpha=0.3)

# 3. Scatter: Transactions vs Amount
for status, color in [('Existing Customer', 'green'), ('Attrited Customer', 'red')]:
    data = df[df['Attrition_Flag']==status]
    axes[1, 0].scatter(data['Total_Trans_Ct'], data['Total_Trans_Amt'], 
                      label=status, alpha=0.5, s=20, c=color)
axes[1, 0].set_xlabel('Transaction Count', fontweight='bold')
axes[1, 0].set_ylabel('Transaction Amount ($)', fontweight='bold')
axes[1, 0].set_title('Transaction Count vs Amount', fontweight='bold', fontsize=13)
axes[1, 0].legend()
axes[1, 0].grid(alpha=0.3)

# 4. Transaction Change Q4 vs Q1
axes[1, 1].hist([df[df['Attrition_Flag']=='Existing Customer']['Total_Amt_Chng_Q4_Q1'],
                 df[df['Attrition_Flag']=='Attrited Customer']['Total_Amt_Chng_Q4_Q1']],
                label=['Retained', 'Churned'], bins=30, color=['green', 'red'], alpha=0.7)
axes[1, 1].axvline(x=0.7, color='blue', linestyle='--', linewidth=2, label='Declining Threshold')
axes[1, 1].set_xlabel('Transaction Change (Q4/Q1 Ratio)', fontweight='bold')
axes[1, 1].set_ylabel('Frequency', fontweight='bold')
axes[1, 1].set_title('Transaction Momentum (Q4 vs Q1)', fontweight='bold', fontsize=13)
axes[1, 1].legend()
axes[1, 1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Statistics
print("\nTRANSACTION STATISTICS:")
print(f"\n{'Metric':<30} {'Churned':>15} {'Retained':>15} {'Difference':>15}")
print("-" * 80)

trans_ct_churn = df[df['Attrition_Flag']=='Attrited Customer']['Total_Trans_Ct'].mean()
trans_ct_retain = df[df['Attrition_Flag']=='Existing Customer']['Total_Trans_Ct'].mean()
print(f"{'Avg Transaction Count':<30} {trans_ct_churn:>15.1f} {trans_ct_retain:>15.1f} {trans_ct_retain-trans_ct_churn:>15.1f}")

trans_amt_churn = df[df['Attrition_Flag']=='Attrited Customer']['Total_Trans_Amt'].mean()
trans_amt_retain = df[df['Attrition_Flag']=='Existing Customer']['Total_Trans_Amt'].mean()
print(f"{'Avg Transaction Amount ($)':<30} {trans_amt_churn:>15.1f} {trans_amt_retain:>15.1f} {trans_amt_retain-trans_amt_churn:>15.1f}")

# Count customers below threshold
low_trans = (df['Total_Trans_Ct'] < 50).sum()
low_trans_pct = (low_trans / len(df)) * 100
print("\n")
print(f"{low_trans:,} customers ({low_trans_pct:.1f}%) have <50 transactions. Therefore, this group has significantly higher churn risk.")
```
